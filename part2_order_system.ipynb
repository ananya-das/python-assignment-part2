## Task 1 — Explore the Menu

```python
# Menu grouped by category
categories = set(dish["category"] for dish in menu.values())

for i in categories:
    print(f"\n===== {i} =====")
    
    for item, details in menu.items():
        if details["category"] == i:
            status = "Available" if details["available"] else "Unavailable"
            print(f"{item:<15} ₹{details['price']:<7} [{status}]")
```

```python
# Total number of items in the menu:
print(f" Total Items in the menu: {len(menu)}")

# Total available Items:
availableItems = sum(1 for i in menu.values() if i["available"])
print(f" Total number of Available Items: {availableItems}")

# The most expensive item (name + price)
expensive_Item = max(menu.items(), key=lambda item: item[1]["price"])
print(f" The most expensive Item in the menu: {expensive_Item[0]} for ₹ {expensive_Item[1]["price"]}")

# All items priced under ₹150 (name + price)
print("\nItems under ₹150: ")
required_Items = [name for name in menu if menu[name]["price"] < 150]
for i in required_Items:
    print(f" {i:<15} : ₹ {menu[i]["price"]}")
```

## Task 2 — Cart Operations

### Add Item to cart

```python
cart = []
```

```python

def addItem(item, count):
    # Check if item is in menu:
    if item not in menu:
        print(f" {item} NOT found in menu!")
        return

    # Check if item is available
    if not menu[item]["available"]:
        print(f" {item} NOT available!")
        return

    # Update cart: If item in cart, increase value
    for i in cart:
        if i["item"] == item:
            i["quantity"] += count
            print(f" {item} added to cart!")
            return
    
    # Add new item
    cart.append({
        "item": item,
        "quantity": count,
        "price": menu[item]["price"]
    })
    print(f" {item} added to cart!")

    
```

### Remove item from cart

```python
def removeFromCart(name):
    for i in cart:
        if i["item"] == name:
            cart.remove(i)
            print(f"Item Removed: {name}")
            return
    print(f"{name} not in cart")
```

```python
def printCart():
    print("\nCurrent Cart:")
    for i in cart:      
        print(f" {i['item']} x {i['quantity']} for ₹{i['price']} each")
```

Adding Items to cart in sequence:

### Final Order Summary:

```python
#Calculations:

total = sum(c["quantity"] * c["price"] for c in cart)
gst = round(total * 0.05, 2)
Grandtotal = total + gst

print("\n========== Order Summary ==========")
for c in cart:
    item_total = c["quantity"] * c["price"]
    print(f"{c['item']:<18} x{c['quantity']}   ₹{item_total:.2f}")

print("-" * 36)
print(f"Subtotal:              ₹{total:.2f}")
print(f"GST (5%):              ₹{gst:.2f}")
print(f"Total Payable:         ₹{Grandtotal:.2f}")
print("====================================")
```

## Task 3 — Inventory Tracker with Deep Copy 

```python
import copy

inventory_backup = copy.deepcopy(inventory)
```

```python
# Modify original to prove deep copy
inventory["Butter Chicken"]["stock"] = 20

print("After Modification:")
print("Inventory:", inventory)
print("Backup   :", inventory_backup)

```

```python
# Restore original inventory
inventory = copy.deepcopy(inventory_backup)
```

```python
for i in cart:
    item = i["item"]
    qty_needed = i["quantity"]
    
    if item in inventory:
        stock = inventory[item]["stock"]
        
        if stock >= qty_needed:
            inventory[item]["stock"] -= qty_needed
        else:
            print(f"⚠ Only {stock} {item} available, reducing all stock to 0")
            inventory[item]["stock"] = 0
            
```

```python
print("\nReorder Alerts:")
for item, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {data['stock']} unit(s) left (reorder level: {data['reorder_level']})")
```

```python
print("\nFinal Inventory:\n", inventory)
print("Backup Inventory:\n", inventory_backup)
```

## Task 4 — Daily Sales Log Analysis 

```python
# Total Revenue per day:

revenue_per_day = {}
   
print("\nRevenue Per Day:")
print("-------------------")
for date, orders in sales_log.items():
    total = sum(order["total"] for order in orders)
    revenue_per_day[date] = total
    print(f"{date} : ₹{total:.2f}")
```

```python
# Best Selling Day:
bestDay = max(revenue_per_day, key=revenue_per_day.get)
print(f" {bestDay} with  ₹{revenue_per_day[bestDay]}")

```

```python
# The most ordered item — the item that appears in the greatest number of individual orders across all days.

ordercount = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            ordercount[item] = ordercount.get(item, 0) + 1

most_ordered = max(ordercount, key=ordercount.get)

print("\nMost Ordered Item:", most_ordered, f"({ordercount[most_ordered]} times)")
```

```python
# Using enumerate, print a numbered list of all orders across all dates 

all_orders = []

for date, orders in sales_log.items():
    for order in orders:
        all_orders.append((date, order))

for i, (date, order) in enumerate(all_orders, start=1):
    items_str = ", ".join(order["items"])
    print(f"{i}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")
```
